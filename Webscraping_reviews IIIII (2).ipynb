```python
from selenium.webdriver import Chrome
import selenium
from selenium import webdriver as wd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re
import pandas as pd
import os
```

```python
chrome_options=wd.ChromeOptions()

driver = wd.Chrome(executable_path="C:\\Users\\akhya\\OneDrive\\Desktop\\Data Management\\Project\\chromedriver.exe",options=chrome_options)
```

```python
link = "https://www.tripadvisor.com/Hotel_Review-g45963-d224853-Reviews-Tuscany_Suites_Casino-Las_Vegas_Nevada.html"
```

```python
review=[]
users = []
stars = []
dateofstay=[]
hotelname = []
for i in range(10):
    driver.get(link)
    rev = driver.find_elements_by_xpath('//q[@class="hotels-review-list-parts-ExpandableReview__reviewText--3oMkH"]')
    user = driver.find_elements_by_xpath('//a[@class="ui_header_link social-member-event-MemberEventOnObjectBlock__member--35-jC"]')
    star = driver.find_elements_by_xpath('//span[@class="ui_bubble_rating bubble_50"]')
    dateofstay1 = driver.find_elements_by_xpath('//span[@class="hotels-review-list-parts-EventDate__event_date--CRXs4"]')
    hotelname1 = driver.find_elements_by_xpath('//h1[@class="hotels-hotel-review-atf-info-parts-Heading__heading--2ZOcD"]')
    for i in rev:
        review.append(i.text)
        hotelname.append(hotelname1[0].text)
    for i in user:
        users.append(i.text)
    for i in range (len(star)):
        a = str(star).split("_")[-1]
        stars.append(a)
    for i in dateofstay1:
        dateofstay.append(i.text)
    next_button = driver.find_element_by_xpath('//a[@class = "ui_button nav next primary " ]')
    link = next_button.get_attribute('href')
```

```python
data_fr21 = pd.DataFrame({'Hotel Name':hotelname,'star':len(star), 'User Name':users,'Date of Stay':dateofstay,'Review':review})
data_fr21.to_csv('Data21.csv')
data_fr21
```
